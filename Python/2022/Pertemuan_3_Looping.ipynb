# Control Flow : Looping

Looping atau perulangan merupakan cara yang digunakan untuk menjalankan perintah yang berulang untuk data berbentuk kelompok seperti list, tuple atau string. Dalam melakukan perintah looping biasanya digunakan perintah For Loop dan While Loop, serta penerapan pernyataan Break dan Continue sebagai pilihan tambahan dalam membuat perintah looping.

## **A. While Loop**

Perintah ‚ÄúWhile‚Äù digunakan dalam melakukan perulangan pada data kelompok yang tidak diketahui jumlah pasti iterasinya. Oleh karena itu, biasanya perintah while akan terus melakukan perulangan selama suatu kondisi masih terpenuhi. Selain itu, pada looping juga kita bisa tambahkan kondisi ‚Äúelse‚Äù seperti layaknya pada Conditional Statement pada bab sebelumnya.



```python
# [blok utama]

# while [boolean] : 
    # [blok dalam while loop]

# [blok utama]
```

Jika boolean bernilai **True** maka blok di dalam blok while akan terus dijalankan secara berulang hingga boolean bernilai **False** atau terjadi interupsi pada kode program (**break** dan **continue**)

```python
# Invinity loop : boolean selalu bernilai True
print("blok utama program")

while True:
    print("invinity loop")

print("blok utama") # program tidak akan sampai ke baris ini
```

**while loop dengan kondisi**

Perulangan harus dibuat agar dapat berhenti secara otomatis. Untuk itu, daripada memasukkan nilai **True/False** secara langsung setelah keyword **while**, kita juga dapat menggunakan operasi" yang menghasilkan tipedata boolean, seperti **Operasi Perbandingan** dan **Operasi Logika** 

```python
# Operasi Perbandingan : < > == != <= >= 

# contoh 1
print("## Blok Utama ##")

x = 0
while x <= 4:
    print("nilai x sekarang : ",x)
    x = x+1 # setara x += 1 ; sebagai increment; kondisi agar while dapat menjadi False

print("## Blok Utama ##") 
 
```

```python
# Operasi Perbandingan : < > == != <= >= 

# contoh 2
print("## Blok Utama ##")

x = 10
while x >= 4:
    print("nilai x sekarang : ",x)
    x = x-1 # setara x += 1 ; sebagai increment; kondisi agar while dapat menjadi False

print("## Blok Utama ##")  
```

```python
# Operasi Logika : and or not 

# contoh 1
print("## Blok Utama ##")

x = 10
buah = "mangga"
while (x > 4) and (buah == "mangga"):
    print("nilai x sekarang : ",x)
    x = x-1 # setara x += 1 ; sebagai increment; kondisi agar while dapat menjadi False

print("## Blok Utama ##")  
```

```python
# Operasi Logika : and or not 

# contoh 2
print("## Blok Utama ##")

x = 6
buah = "mangga"
while (x > 4) or (buah == "mangga"):
    print("nilai x sekarang : ",x)
    x = x-1 # setara x += 1 ; sebagai increment; kondisi agar while dapat menjadi False
    if x == 1:
        buah = "apel" 

print("## Blok Utama ##")  
```

```python
# Operasi Logika : and or not 

# contoh 3
print("## Blok Utama ##")

x = 6
buah = "mangga"
while (x > 4) or (not buah == "mangga"): # "not" membalikkan nilai boolean dari "False" ke "True" dan sebaliknya
    print("nilai x sekarang : ",x)
    x = x-1 # setara x += 1 ; sebagai increment; kondisi agar while dapat menjadi False

print("## Blok Utama ##")  
```

## **B. For Loop**

Perintah ‚ÄúFor‚Äù digunakan dalam melakukan perulangan pada data kelompok yang sudah diketahui jumlah iterasinya dan akan berhenti jika iterasinya telah dieksekusi semuanya.


```python
# blok utama program

# for [var] in [collection] :
     # [blok dalam perulangan for] 

# blok utama
```

Contoh collection yang dapat digunakan pada perulangan "for" adalah "range" dan "list"

**Range**

range (rentang) adalah salah satu tipedata yang berupa kumpulan bilangan integer berurut dalam rentang yang ditentukan

range(start, end, step)

> start: batas mula" 
> end : sebagai batas akhir (rentang akan berhenti tepat sebelum end)
> step : langkah/lompatan angka atau selisih angka dalam rentang  

```python
# jika parameter (e) hanya 1 maka rentang bermula dari 0 hingga e-1
x = range(10)
print([*x]) # abaikan "[* ]"; hanya untuk menampilkan range ke dalam list
```

```python
# dua parameter (s,e) maka rentang bermula dari s hingga e
x = range(2,10)
print([*x])
```

```python
# tiga parameter (s,e,st) maka rentang bermula dari s hingga e dengan selisih tiap komponen sebanyak st angka
x = range(1,10,2)
print([*x])
```

**Perulangan dengan range**

Perulangan for paling dasar pada python adalah dengan menggunakan range()  

```python
# contoh 1
for i in range(4) : 
    print(i)
```

```python
hasil = 0
for i in range(1,10, 2) :
    hasil += i # setara " hasil = hasil + i "
    print("unsur : ", i)

print("total : ", hasil)
```

**List**

List adalah salah satu tipedata yang berupa sekumpulan data" tertentu. anggotanya dapat berupa int, float, string, boolean, bahkan list itu sendiri

```python
x = [10, 3.14, "mangga", ["jaka", "joko", "jaki"]]
print(x)
```

```python
# mengakses komponen list (index dimulai dari 1)
x = [10, 3.14, "mangga", ["jaka", "joko", "jaki"]]
print( x[0] )
```

```python
# mengakses komponen list di dalam list
x = [10, 3.14, "mangga", ["jaka", "joko", "jaki"]]
print( x[3][2] )
```

**Perulangan for untuk mengakses data dalam collection**

```python
# mengakses unsur menggunakan index
x = [10, 3.14, "mangga", ["jaka", "joko", "jaki"]]
for i in range(x.__len__()) : # x.__len__() mengembalikan panjang / banyak unsur dari collection
    print( x[i] )
```

```python
# mengakses unsur secara langsung 
x = [10, 3.14, "mangga", ["jaka", "joko", "jaki"]]
for i in x:
    print(i)
```

```python
# mengakses list di dalam list (nested loop)
daftar = [["budi", 19, "bola"], ["tono", 14, "renang"], ["rehan", 16, "voly"]] 
for orang in daftar :
    for keterangan in orang:
        print(keterangan, end=", ")
    print()
```

## **C. Break and Continue**

Penggunaan ‚ÄúBreak‚Äù dan ‚ÄúContinue‚Äù digunakan untuk menambahkan kondisi perintah pada looping sehingga dalam looping tersebut akan ada kondisi ‚ÄúIf‚Äù yang ditambahkan. Break digunakan jika kita ingin menghentikan sebuah looping jika suatu kondisi telah terpenuhi, sedangkan continue digunakan untuk melewati iterasi sekarang dan melanjutkan langsung ke iterasi selanjutnya jika kondisinya terpenuhi.

```python
# break (menghentikan secara paksa / keluar dari blok perulangan)
for i in range(10) :
    print(i)
    if i == 2:
        break
```

```python
# continue (melangkahi, melompat ke perulangan selanjutnya)
for i in range(4):
    if i == 2:
        continue
    print(i)
```

```python
# continue (melangkahi, melompat ke perulangan selanjutnya)
for i in range(10):
    if i > 3 and i <=8 :
        continue
    print(i)
```

## **D. Try Except**

Ketika terjadi kesalahan, atau exception, Python biasanya akan berhenti dan menghasilkan pesan kesalahan. Pengecualian ini dapat ditangani menggunakan pernyataan try:

```python
x = "5"
hasil = 0
try :                       # berfungsi sebagai tempat menjalankan perintah yang berpotensi error
    hasil = x % 2 # error
except :                    # blok ini akan dijalankan jika pada blok "try" terjadi error
    hasil = int(x) * 1000   #dijalankan
else :                      # blok ini akan dijalankan jika pada blok "try" TIDAK terjadi error
    print("tidak terjadi error")
finally :                   # blok ini akan tetap dijalankan ketika blok "try" terjadi error maupun tidak error 
    print(hasil)
    print("Program berhasil dijalankan") 
```
