# **A. Conditional Statement**
Conditional statement (yang dalam bahasa indonesia disebut pernyataan bersyarat atau pengkondisian) adalah fitur dari bahasa pemrograman yang melakukan perhitungan atau tindakan yang berbeda tergantung pada kondisi yang ditentukan programmer. Pengkondisian membuat program memiliki cabang yang masing-masing akan tereksekusi jika persyaratannya terpenuhi.

bahasa pemrograman seperti C, C++, dan Java setidaknya memiliki fitur conditional statement seperti if statement dan switch case, namun dalam bahasa python switch case baru di adaptasi pada versi 3.10 keatas yang dikenal sebagai ‚Äústructural pattern matching‚Äù atau ‚Äúmatch case‚Äù. Artinya, pada versi sebelumnya pengkondisian hanya dapat dilakukan menggunakan if statement. Ekspresi terdiri dari satu atau beberapa operator perbandingan dan operator logika yang akan menghasilkan nilai True (benar) atau False (salah).

# **C. 'if' Statement**
If statement adalah jenis pengkondisian yang paling mendasar, di mana kode dieksekusi apabila ekspresi terpenuhi atau bernilai True (benar). Pernyataan dari if statement harus memiliki indent minimal sepanjang satu spasi di awal tiap baris kode.Suatu pernyataan dapat berupa satu baris atau satu blok kode. 

>Modul PP.drawio.png

```python
kondisi = True

if (kondisi): # 'if' merupakan keyword untuk membuat conditional statement. setelah if merupakan boolean 
              #      dapat berupa operator (perbandingan/logika), variabel, langsung dengan nilai 'True' atau 'False'
  print("blok if ter-eksekusi") # blok ini di eksekusi
# üëÜ bagian awal baris agak maju (memiliki indent/spasi)
print("blok utama ter-eksekusi") 
```

```python
kondisi = False

if (kondisi):
  print("blok if ter-eksekusi") # blok blok ini tidak dieksekusi
# üëÜ bagian awal baris agak maju (memiliki indent/spasi)
print("blok utama ter-eksekusi") 
```

**Menggunakan Beberapa 'if' Statement**

```python
angka = 1000

if angka < 10:
  print(angka, "adalah satuan")

if angka < 100:
  print(angka, "adalah puluhan")

if angka < 1000:
  print(angka, "adalah ratusan")

if angka < 10_000:
  print(angka, "adalah ribuan")
```

# **D. 'if-else' Statement**
Conditional ‚Äúif-else‚Äù digunakan ketika bagian benar dan salah dari kondisi tertentu ditentukan untuk dieksekusi. Ketika kondisinya benar, pernyataan di dalam blok 'if' dieksekusi; jika kondisinya salah, program pada blok 'else' akan dieksekusi.

> Modul PP.drawio (6).png

```python
kondisi = False

if (kondisi):
  print("blok if ter-eksekusi")
  
else : 
  print("blok else ter-eksekusi")

print("blok utama ter-eksekusi") 
```

```python
# Contoh 1
angka = 5

if angka >= 0:
    print("bilangan positif atau nol")

else:
     print(angka, "adalah bilangan negatif")
```

```python
# Contoh 2
x = 5
y = 20

if x == y:
    print("x dan y bernilai sama")

else:
     print("x tidak sama dengan y")
```

# **E. 'if-elif-else' statement**
Pernyataan elif memungkinkan Anda untuk memeriksa beberapa ekspresi dan mengeksekusi blok kode segera setelah salah satu kondisi mengevaluasi ke True. Dalam hal ini, kondisi if dievaluasi terlebih dahulu. Jika salah, pernyataan elif akan dieksekusi, jika itu juga salah, pernyataan else akan dieksekusi. 

> Modul PP.drawio (7).png

```python
kondisi_if = False
kondisi_elif = True

if (kondisi_if):
  print("blok if ter-eksekusi")

elif (kondisi_elif):
  print("blok elif ter-eksekusi")
  
else : 
  print("blok else ter-eksekusi")

print("blok utama ter-eksekusi") 
```

```python
# Contoh 1
angka = 0

if angka > 0:
    print(angka, "adalah bilangan positif")
    
elif angka == 0:
    print(angka, "adalah bilangan nol")

else:
     print(angka, "adalah bilangan negatif")
```

# **F. Nested IF Statement**
Pernyataan 'if' bersarang adalah pernyataan dimana terdapat pernyataan 'if' di dalam pernyataan 'if' lainnya. Ini digunakan ketika variabel harus diproses lebih dari sekali. Pernyataan 'if', 'if-else', dan 'if‚Ä¶elif‚Ä¶else' dapat digunakan dalam blok 'if', 'elif', 'else' untuk membuat kondisi bersarang tanpa batas jumlah tumpukan.

>Modul PP.drawio (8).png

```python
kondisi_luar = True
kondisi_dalam = True

if (kondisi_luar):
  print("blok if 'luar' ter-eksekusi")
  
  if (kondisi_dalam):
    print("blok if 'dalam' ter-eksekusi")

else : 
  print("blok else ter-eksekusi")

print("blok utama ter-eksekusi") 
```

```python
# Contoh 1
angka = 0

if angka >= 0:
    print(angka, "adalah bilangan positif")
    if angka == 0:
        print(angka, "adalah bilangan nol")
else:
     print(angka, "adalah bilangan negatif")
```

```python
# Contoh 2
angka = 4

if angka >= 0:
    print(angka, "adalah bilangan positif")

    if (angka % 2) == 0:
        print(angka, "adalah bilangan genap positif")
else:
     print(angka, "adalah bilangan negatif")
```

# **H. Shorthand 'if-else'**
Shorthand if-else digunakan untuk menyebutkan pernyataan If-else dalam satu baris di mana hanya ada satu pernyataan untuk dieksekusi di blok if dan else.


```python
kondisi = False
print("blok if ter-eksekusi") if kondisi else print("blok else ter-eksekusi") 
```

```python
a = 123
print(a, " besar seratus")if a > 100 else print(a, " kecil seratus")
```

# **Contoh Kasus**

Buatlah program yang dapat menecek kelulusan siswa berdasarkan aturan berikut:
* Mengerjakan tugas Mid dan Kehadiran >= 80% dari 16 pertemuan dan nilai rata" > 70
* Mengerjakan tugas Mid Kehadiran >= 60% dari 16 pertemuan dan nilai rata" > 85
* Tidak Mengerjakan tugas Mid dan Kehadiran >= 60% dan mengikuti lomba dan nilai rata" > 60

>Format input. terdapat empat inputan yaitu:
* nama : string 
* tugas_mid : string ("ya"/"tidak")
* kehadiran : int (antara 0 - 16)
* nilai_avg : float (antara 0 - 100)
* lomba : string ("ya" / "tidak")

>Format output : 

>jika input valid : cetak "[nama siswa] dinyatakan [lulus/tidak]"

>jika input tidak valid : cetak pemberitahuan input yang salah cth: "kehadiran harus dalam rentang 0 - 16" 







```python
# membuat variable untuk menampung nilai input
nama = input("Nama siswa : ")
tugas_mid = input("Tugas mid (ya/tidak) : ")
kehadiran = int(input("Kehadiran (0 - 16) : "))
nilai_avg = float(input("Nilai rata-rata (0 - 100) : ")) 
lomba = input("Mengikuti lomba (ya/tidak) : ")

print("-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-")

# konversi nilai kehadiran kedalam persentase
prst_kehadiran = kehadiran/16 * 100

# validasi inputan
tugas_mid_valid = (tugas_mid == "ya" or tugas_mid == "tidak")
kehadiran_valid = (kehadiran >= 0 and kehadiran <= 16)
nilai_avg_valid = (nilai_avg >= 0 and nilai_avg <= 100)
lomba_valid = (lomba == "ya" or lomba == "tidak")

input_valid = tugas_mid_valid and kehadiran_valid and nilai_avg_valid and lomba_valid

if input_valid:

  if (tugas_mid == "ya") and (prst_kehadiran >= 80) and (nilai_avg > 70): # cek kondisi lulus ke-1
    print("\n", nama, "dinyatakan lulus!")

  elif (tugas_mid == "ya") and (prst_kehadiran >= 60) and (nilai_avg > 85): # cek kondisi lulus ke-2
    print("\n", nama, "dinyatakan lulus!")

  elif (prst_kehadiran >= 60) and (nilai_avg > 65) and (lomba == "ya"): # cek kondisi lulus ke-3
    print("\n", nama, "dinyatakan lulus!")

  else:
    print("\n", nama, "dinyatakan tidak lulus!")

else :
  print("\nError: Input tidak valid")

  if not tugas_mid_valid:
    print("Tugas mid harus antara 'ya' dan 'tidak'" )
  
  if not kehadiran_valid:
    print("Kehadiran harus dalam rentang 0 - 16")
  
  if not nilai_avg_valid:
    print("Nilai avg harus dalam rentang 0 - 100")

  if not lomba_valid:
    print("Lomba harus antara 'ya' dan 'tidak'")

  
```
